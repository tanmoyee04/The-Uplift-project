```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Line Graph
apples=[83,33,59,10,66]
years=[2015,2016,2017,2018,2019]
plt.xlabel('no.of years')
plt.ylabel('no.of lables')
plt.plot(years,apples)
```

```python
#Multiline graph
apples=[83,33,59,10,66]
years=[2015,2016,2017,2018,2019]
oranges=[58,10,41,98,22]
plt.xlabel('no.of years')
plt.ylabel('no.of lables')
plt.plot(years,apples)
plt.plot(years,oranges)
plt.title('Multiline Graph')
plt.legend(['apples','orange'])
```

```python
apples=[83,33,59,10,66]
years=[2015,2016,2017,2018,2019]
oranges=[58,10,41,98,22]
plt.xlabel('no.of years')
plt.ylabel('no.of lables')
plt.plot(years,apples,marker='*')
plt.plot(years,oranges,marker='^')
plt.title('Multiline Graph')
plt.legend(['apples','orange'])
```

```python
apples=[83,33,59,10,66]
years=[2015,2016,2017,2018,2019]
oranges=[58,10,41,98,22]
plt.xlabel('no.of years')
plt.ylabel('no.of lables')
plt.plot(years,apples,'or',color='green')
plt.plot(years,oranges,'or')
plt.title('Multiline Graph')
plt.legend(['apples','orange'])
```

```python
sns.set_style('darkgrid')
apples=[83,33,59,10,66]
years=[2015,2016,2017,2018,2019]
oranges=[58,10,41,98,22]
plt.xlabel('no.of years')
plt.ylabel('no.of lables')
plt.plot(years,apples,marker='*')
plt.plot(years,oranges,marker='^')
plt.title('Multiline Graph')
plt.legend(['apples','orange'])
```
